### NER

```python
from flair.data import Sentence
from flair.models import MultiTagger
from flair.models import SequenceTagger
from segtok.segmenter import split_single

# load the NER tagger
# tagger = SequenceTagger.load('ner')

# load tagger for POS and NER 
tagger: MultiTagger = MultiTagger.load(['pos', 'ner'])
```

```python
small_df = master_df[:10]
```

```python
def extract_entities_title(title):
    sentence = Sentence(title)
    tagger.predict(sentence)
    return sentence.to_dict(tag_type='ner')
```

```python
def extract_entities_content(content, keyword):
    all_entities = []
    sentences = [Sentence(sent, use_tokenizer=True) for sent in split_single(content) if keyword in sent]
    tagger.predict(sentences)
    for sentence in sentences:
        all_entities.append(sentence.to_dict(tag_type='ner'))
    return all_entities
```

```python
def extract_entities_content_only_first_sentence(content, keyword):
    sentences = [sent for sent in split_single(content) if keyword in sent]
    first_sent = Sentence(sentences[0], use_tokenizer=True)
    tagger.predict(first_sent)
    
    return first_sent.to_dict(tag_type='ner')
```

```python
def extract_companies(entity_dict):
    companies = []
    for entity in entity_dict['entities']:
        if entity['labels'][0].value == "ORG" and entity['labels'][0].score > 0.95:
            companies.append(entity['text'])
    
    return list(set(companies))

def extract_all_companies(all_sentences):    
    all_companies = []
    for sentence in all_sentences:
        all_companies.append(extract_companies(sentence))
        
    flat_list = [item for sublist in all_companies for item in sublist]
    
    return list(set(flat_list))
```

```python
small_df['all_entities_title'] = small_df['title'].apply(extract_entities_title)
```

```python
small_df['all_entities_content'] = small_df.apply(lambda x: extract_entities_content(x['content'], x['keywords']), axis = 1)
```

```python
small_df['first_sent_content'] = small_df.apply(lambda x: extract_entities_content_only_first_sentence(x['content'], x['keywords']), axis = 1)
```

```python
small_df['title_companies'] = small_df['all_entities_title'].apply(extract_companies)
```

```python
small_df['content_companies'] = small_df['all_entities_content'].apply(extract_all_companies)
```

```python
small_df['first_sent_companies'] = small_df['first_sent_content'].apply(extract_companies)
```

```python
def rule1(text):
    
    doc = nlp(text)
    
    sent = []
    
    for token in doc:
        
        # if the token is a verb
        if (token.pos_=='VERB'):
            
            phrase =''
            
            # only extract noun or pronoun subjects
            for sub_tok in token.lefts:
                
                if (sub_tok.dep_ in ['nsubj','nsubjpass']) and (sub_tok.pos_ in ['NOUN','PROPN','PRON']):
                    
                    # add subject to the phrase
                    phrase += sub_tok.text

                    # save the root of the verb in phrase
                    phrase += ' '+token.lemma_ 

                    # check for noun or pronoun direct objects
                    for sub_tok in token.rights:
                        
                        # save the object in the phrase
                        if (sub_tok.dep_ in ['dobj']) and (sub_tok.pos_ in ['NOUN','PROPN']):
                                    
                            phrase += ' '+sub_tok.text
                            sent.append(phrase)
            
    return sent
```
