### Average questions

```python
sentence_df['general_questions'] = sentence_df['general_questions'].apply(lambda x: ast.literal_eval(x))
```

```python
data = sentence_df[~sentence_df['head_questions'].isna()]; data
data.reset_index(drop = True, inplace = True)
```

```python
data['head_questions'] = data['head_questions'].apply(lambda x: ast.literal_eval(x))
data['tail_questions'] = data['tail_questions'].apply(lambda x: ast.literal_eval(x))
```

```python
data['head_before'] = data['head_questions'].apply(lambda x: x['before'])
data['head_after'] = data['head_questions'].apply(lambda x: x['after'])

data['tail_before'] = data['tail_questions'].apply(lambda x: x['before'])
data['tail_after'] = data['tail_questions'].apply(lambda x: x['after'])
```

```python
data
```

```python
# calculate length of head questions and tail questions before and after (the average probably)
print(mean([len(x) for x in list(data['head_before'])]))
print(max([len(x) for x in list(data['head_before'])]))
print(min([len(x) for x in list(data['head_before'])]))
print('-----------------------------------')
print(mean([len(x) for x in list(data['head_after'])]))
print(max([len(x) for x in list(data['head_after'])]))
print(min([len(x) for x in list(data['head_after'])]))
print('-----------------------------------')
print(mean([len(x) for x in list(data['tail_before'])]))
print(max([len(x) for x in list(data['tail_before'])]))
print(min([len(x) for x in list(data['tail_before'])]))
print('-----------------------------------')
print(mean([len(x) for x in list(data['tail_after'])]))
print(max([len(x) for x in list(data['tail_after'])]))
print(min([len(x) for x in list(data['tail_after'])]))
```
